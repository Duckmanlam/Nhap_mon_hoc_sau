```python
import pandas as pd
#Câu 1: Tạo dataframe từ dữ liệu
data = {
    'Product': ['Laptop', 'Phone', 'Chair', 'Desk', 'Headphone', 'Monitor', 'Sofa', 'Keyboard', 'Mouse', 'Tablet'],
    'Category': ['Electronics', 'Electronics', 'Furniture', 'Furniture', 'Accessories', 'Electronics', 'Furniture', 'Accessories', 'Accessories', 'Electronics'],
    'Price': [1500, 800, 120, 300, 200, 600, 700, 150, 50, 400],
    'Date': ['2024-01-15', '2024-02-10', '2024-03-15', '2024-03-18', '2024-04-12', '2024-05-20', '2024-06-20', '2024-02-08', '2024-07-14', '2024-09-30'],
    'Quantity': [2, 5, 3, 1, 4, 2, 1, 6, 8, 3]
}

df = pd.DataFrame(data)
# Chuyển đổi cột Date sang định dạng datetime
df['Date'] = pd.to_datetime(df['Date'])

```

```python
# Câu 2: Hiển thị thông tin cơ bản của dataframe
print("Thông tin Dataframe:")
print(df.info())
```

```python
# # Câu 3:  Tính giá trị trung bình của cột Price
price_mean = df['Price'].mean()
# category_mean = df.groupby('Category')['Price'].mean()
print("\nGiá trị trung bình của Price theo Category:")
print(price_mean)
```

```python
#Câu 4:  Nhóm dữ liệu theo Category và tính trung bình của Quantity
category_quantity_mean = df.groupby('Category')['Quantity'].mean()
print("\nGiá trị trung bình của Quantity theo Category:")
print(category_quantity_mean)
```

```python
#Câu:5 Sắp xếp dataframe theo Price giảm dần
df_sorted = df.sort_values(by='Price', ascending=False)
print("\nDataframe sau khi sắp xếp theo Price giảm dần:")
print(df_sorted)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
#6. Vẽ biểu đồ cột cho Price:
plt.figure(figsize=(10, 6))
sns.barplot(x='Product', y='Price', data=df)
plt.title('Giá của các sản phẩm')
plt.xlabel('Sản phẩm')
plt.ylabel('Giá (USD)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#7 7. Vẽ biểu đồ phân tán cho <Quantity> và <Price>.
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Quantity', y='Price', data=df, hue='Category', size='Quantity', sizes=(50, 200))
plt.title('Mối quan hệ giữa Số lượng và Giá sản phẩm')
plt.xlabel('Số lượng')
plt.ylabel('Giá (USD)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Danh mục')
plt.tight_layout()
plt.show()
```

```python
#8 Sắp xếp dữ liệu theo ngày tháng
df_sorted_date = df.sort_values('Date')

plt.figure(figsize=(12, 6))
plt.plot(df_sorted_date['Date'], df_sorted_date['Price'], marker='o', linestyle='-', linewidth=2)

for i, row in df_sorted_date.iterrows():
    plt.annotate(row['Product'], (row['Date'], row['Price']), 
                 textcoords="offset points", xytext=(0,10), ha='center')

plt.title('Biến động giá theo thời gian')
plt.xlabel('Ngày')
plt.ylabel('Giá (USD)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Câu 9 :  Vẽ biểu đồ cột thể hiện tổng doanh thu của từng Category
category_revenue = df.groupby('Category')['Price'].sum()
plt.figure(figsize=(10, 6))
sns.barplot(x=category_revenue.index, y=category_revenue.values)
plt.title('Doanh thu theo Category')
plt.xlabel('Category')
plt.ylabel('Doanh thu (USD)')
plt.tight_layout()
plt.show()
```

```python
#Câu 10: Kiểm tra giá trị NaN trong DataFrame
nan_check = df.isna().sum()
print("Kiểm tra giá trị NaN trong DataFrame:")
print(nan_check)

if nan_check.sum() == 0:
    print("Không có giá trị NaN nào trong DataFrame.")
else:
    print(f"Có tổng cộng {nan_check.sum()} giá trị NaN trong DataFrame.")
```

```python
#Câu 11 Kiểm tra các giá trị trùng lặp dựa trên cột Product
duplicates = df[df.duplicated(subset=['Product'], keep=False)]
print("Các hàng có giá trị Product trùng lặp:")
print(duplicates)

if duplicates.empty:
    print("Không có giá trị Product nào trùng lặp.")
else:
    # Loại bỏ các hàng trùng lặp, giữ lại hàng đầu tiên
    df_no_duplicates = df.drop_duplicates(subset=['Product'], keep='first')
    print("\nDataFrame sau khi loại bỏ các hàng trùng lặp:")
    print(df_no_duplicates)
```

```python
#Câu 12 Lọc các đơn hàng có Quantity > 5
large_orders = df[df['Quantity'] > 5]
print("Các đơn hàng có số lượng lớn hơn 5:")
print(large_orders)
```

```python
#Câu 13 Tính tổng doanh thu
df['Revenue'] = df['Price'] * df['Quantity']
total_revenue = df['Revenue'].sum()

print(f"Tổng doanh thu: {total_revenue} USD")

# Tính doanh thu theo từng danh mục
category_revenue = df.groupby('Category')['Revenue'].sum()
print("\nDoanh thu theo từng danh mục:")
print(category_revenue)

# Tính phần trăm đóng góp của từng danh mục
category_percentage = (category_revenue / total_revenue * 100).round(2)
print("\nPhần trăm đóng góp của từng danh mục:")
print(category_percentage)
```

```python
#Câu 14 Thêm cột Sales_Level

def get_sales_level(revenue):
    if revenue < 500:
        return 'Low'
    elif 500 <= revenue <= 2000:
        return 'Medium'
    else:  # revenue > 2000
        return 'High'

df['Sales_Level'] = df['Revenue'].apply(get_sales_level)
print("\nDataFrame sau khi thêm cột Sales_Level:")
print(df[['Product', 'Revenue', 'Sales_Level']])
sales_level_counts = df['Sales_Level'].value_counts().sort_index()

```

```python
#Câu 15 Vẽ biểu đồ cột thể hiện số lượng đơn hàng theo mức
sales_level_counts = df['Sales_Level'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
ax = sns.countplot(x='Sales_Level', data=df, order=['Low', 'Medium', 'High'])
plt.title('Số lượng đơn hàng theo mức doanh thu')
plt.xlabel('Mức doanh thu')
plt.ylabel('Số lượng đơn hàng')

# Thêm số lượng cụ thể trên mỗi cột
for i, count in enumerate(sales_level_counts):
    ax.text(i, count + 0.1, str(count), ha='center')

plt.tight_layout()
plt.show()
```

```python
#Câu 16 Thêm cột Discount dựa trên điều kiện Revenue
def get_discount_rate(revenue):
    if revenue > 1000:
        return 0.10  # 10%
    else:
        return 0.05  # 5%

df['Discount'] = df['Revenue'].apply(get_discount_rate)
print("\nDataFrame sau khi thêm cột Discount:")
print(df[['Product', 'Revenue', 'Discount']])

# Câu 17 Tính doanh thu sau giảm giá và tạo cột Final_Revenue
df['Final_Revenue'] = df['Revenue'] - (df['Revenue'] * df['Discount'])

# Hiển thị kết quả
print("DataFrame sau khi thêm cột Final_Revenue:")
print(df[['Product', 'Revenue', 'Discount', 'Final_Revenue']])
```

```python
#Câu 18 Tạo cột Month từ cột Date
df['Month'] = df['Date'].dt.month
print("\nDataFrame sau khi thêm cột Month:")
print(df[['Product', 'Date', 'Month']])
#Câu 19 vẽ biểu đồ đường thể hiện doanh thu theo tháng
monthly_revenue = df.groupby('Month')['Revenue'].sum().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(monthly_revenue['Month'], monthly_revenue['Revenue'], marker='o', linestyle='-', linewidth=2, color='blue')

# Thêm nhãn dữ liệu trên mỗi điểm
for i, row in monthly_revenue.iterrows():
    plt.annotate(f"{row['Revenue']}", 
                 (row['Month'], row['Revenue']), 
                 textcoords="offset points", 
                 xytext=(0,10), 
                 ha='center')

plt.title('Tổng doanh thu theo tháng')
plt.xlabel('Tháng')
plt.ylabel('Doanh thu (USD)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(range(1, 13))  # Hiển thị đủ 12 tháng
plt.tight_layout()
plt.show()
```
